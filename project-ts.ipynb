```python
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
plt.style.use('bmh')
```

```python
# Download data
# ======================================================================================
url = ('https://raw.githubusercontent.com/vanilsongomes/ts_sunspot/main/data/SN_m_tot_V2.0.csv')
data = pd.read_csv(url,sep=';', header=None)
data.columns = ['year', 'month', 'FracDate', 'Sunspots', 'Std', 'Observations', 'Verified']
data = data[['year', 'month', 'Sunspots']]
df_idx = data.set_index(["year"], drop=True)
```

```python
df_idx.head()
```

```python
#Plot manchas mensais
plt.figure(figsize=(15, 7))
plt.plot(df_idx.Sunspots)
plt.title('Sunspots Monthly')
plt.grid(True)
plt.show()
```
